# Formatos de radar soportados

La codificación binaria que se realiza a los datos de los radares es el mayor obstaculo para muchos usuarios potenciales del radar. Sino fuera poco no se puede acceder facilmente al software decodificador. En caso de que los formatos esten documentados la implementación e un decodificador es un trabajo de programación laborioso. 

Este notebook provee una serie de ejemplos de los formatos más usados y soportados por wradlib. Las librerias estan en constante desarrollo, por lo que puede que existan más formatos soportados.

Se puede hacer uso de librerias como netCDF4 o h5py para codificar los resultados de un análisis realizado.

## Formato Aleman del servicio meteorológio: DX format

El servición climático Alemán usa el formato DX que codifica los barridos del radar local. Los datos DX se encuentran en coordenadas polares y los datos llevan por nombre la siguiente convención:
<br>
    <center>raa00-dx_location-id-YYMMDDHHMM-location-abreviation---bin</center>
   

```python
import numpy as np
import wradlib as wl
import pylab as pl
```

```python
fpath = "/home/aceron/Documentos/GitHub/Tesis/wradlib-data-master/"
fname = "dx/raa00-dx_10908-0806021655-fbg---bin.gz"
data, meta = wl.io.read_dx(fpath+fname)
```

Hasta este momento tenemos en data un arreglo dos dimensional (número de angulos, número de muestras). Esto significa que el número de filas en el arreglo corresponde al número de ángulos del azimuth del barrido del radar, mientras que las columnas corresponden al número de muestras por emisión.

```python
fig = pl.figure(figsize=(10,8))
im = wl.vis.plot_ppi(data, fig=fig, proj='cg')
```

## RADOLAN 

El formati cuantitativo compuesto de DWD (Servicio climático Aleman) fue estbalecido en el curso del projecto RADOLAN. 
La composición de los elementos cuantitativos están distribuidos por una serie de series (e.j. serie-S, serie-R, serie-E)
https://www.dwd.de/DE/leistungen/radolan/radolan_info/radolan_radvor_op_komposit_format_pdf.pdf?__blob=publicationFile&v=5

La composición RADOLAN tiene una resolucióon espacial de 1 km x 1 km, con la composición nacional (serie R,S) hay un mallado de 900 x 900, en la composición europea hay un mallado de 1500x1400. La projección es estereográfica polar. 


```python
fname = "radolan/misc/raa01-rw_10000-1408102050-dwd---bin.gz"
data, meta = wl.io.read_radolan_composite(fpath+fname)
```

```python
fig = pl.figure(figsize=(10,8))
im = wl.vis.plot_ppi(data, fig=fig)
```

```python
#Enmascarando los datos faltantes
maskdata = np.ma.masked_equal(data,meta["nodataflag"])
print(maskdata.shape)
print(data.shape)
```

```python
fig = pl.figure(figsize=(10,8))
# get coordinates
radolan_grid_xy = wl.georef.get_radolan_grid(900,900)
x = radolan_grid_xy[:,:,0]
y = radolan_grid_xy[:,:,1]

# create quick plot with colorbar and title
pl.figure(figsize=(10,8))
pl.pcolormesh(x, y, maskdata)
```

## HDF5

HDF5 es un modelos de datos, libreria y formato de archivo para almacenar y administrar data.

La naturaleza herarquica de los datos DHF5 puede ser descrita de forma similar a los directoriso, archivos y enlaces a un disco duro. Los metadatos son almacenados y llamadas como atributos; esos atributos están organizados juntos y son llamados grups. Los datos binarios son almacenados y llamados dataset. 

La raíz (o la capa de información superior) contiene tres grupos de metadatos: what (objeto, información del modelo, información de los datos y el tiempo), where (información geográfica), who (calidad y recomendaciones).

Para una vista rápida de los datos, existe la entrada datos1 que contiene una matriz que puede ser representada de forma CAPPI

```python
fname = 'hdf5/knmi_polar_volume.h5'
fdata = wl.io.read_opera_hdf5(fpath+fname)
print(len(fdata.keys()))
print(fdata.keys())
```

```python
fig = pl.figure(figsize=(10,8))
im = wl.vis.plot_ppi(fdata['dataset1/data1/data'], fig=fig, proj='cg')
```

## Vaisala Sigmet IRIS

IRIS se refiere al sistema comercial Vaisala Sigmet Interactive Radar Information System. El Vaisala Sigmet Digital Receivers ex = 'sigmet/cor-main131125105503.RAW2049'porta los datos en un formato de archivo extremadamente bien documentado.

Al igual que en el formato HDF5, wradlib transforma la información en un diccionario que puede ser usado por el usuario.

Los datos estan almacenados usando objetos de numpy como numpy.array.

Es común identificar archivos IRIS pues su formato tiene una terminación del tipo *.RAWXXXX

```python
fname = 'sigmet/cor-main131125105503.RAW2049'
fcontent = wl.io.read_iris(fpath+fname)
print(fcontent.keys())
```
